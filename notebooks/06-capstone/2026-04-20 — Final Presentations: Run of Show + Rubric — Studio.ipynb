## Presentation Format

- **Time:** 8 minutes per team, followed by 2 minutes of questions.
- **Format:** Live demo from your notebook, slides optional. If you use slides, keep them minimal — your notebook is the star.
- **Who presents:** Every team member must speak for at least one section.
- **Order:** Randomized. We will draw team numbers at the start of class.

## What to Cover

Your presentation should walk through:

1. **The pipeline (1-2 min):** Show your GCS authentication, downloading data from the bucket, and uploading cleaned data. Prove the cloud pipeline works.

2. **The cleaning (1-2 min):** Show one or two interesting cleaning decisions. The vehicle consolidation is a good one. Show before vs. after.

3. **The analysis (3-4 min):** Pick your 2-3 strongest questions and walk through the findings. Show the charts. Explain what the data says.

4. **The recommendation (1 min):** For Question 5 (connector type investigation), make your recommendation. Should the city invest in CHAdeMO or not? Back it up.

You do not need to cover all 5 questions in the presentation. Pick your strongest work.

## Presentation Tips

- **Do not read your notebook out loud.** The audience can read. Talk about what the data means, not what the code does.
- **Start with the punchline.** Lead with your most interesting finding, then show how you got there.
- **One chart per point.** Do not show 6 charts in 8 minutes. Pick the ones that tell the clearest story.
- **Practice the transitions.** If three people are presenting, practice who hands off to whom.
- **Have a backup.** If the live demo fails (GCS times out, Kaggle is slow), have screenshots of your key outputs.

## Presentation Grading Rubric

The presentation is part of your capstone grade (folded into the overall rubric, not a separate grade).

| Criteria | Excellent | Good | Needs Work |
|----------|-----------|------|------------|
| **Cloud demo** | Live GCS auth + upload shown, works smoothly | Shown but with minor issues | Not demonstrated or pre-recorded only |
| **Cleaning narrative** | Clear before/after, explains trade-offs | Shows cleaning but does not explain why | Skipped or unclear |
| **Analytical depth** | Findings are specific, supported by data, charts are clear | Findings are reasonable but vague | Findings are unsupported or missing |
| **Recommendation quality** | Evidence-based, considers alternatives | Has a recommendation but weak evidence | No recommendation or pure opinion |
| **Delivery** | All members speak, good pacing, engaging | Uneven participation or pacing issues | One person does everything, over time |
